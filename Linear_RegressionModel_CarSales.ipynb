```python
### Changing price to log
```

```python
data_cleaned
```

```python
log_price = np.log(data_cleaned.Price)
```

```python
data_cleaned = data_cleaned.drop(['Model'],axis=1)
data_cleaned = data_cleaned.drop(['Price'],axis=1)
```

```python
data_cleaned["log_price"] = log_price
data_cleaned
```

```python
# By checking multicollinearity it is advised to drop year as it's VIF value is near to 10(>5). 
```

```python
data_multicollinearity = data_cleaned.drop(['Year'],axis=1)
```

```python
data_with_dummies = pd.get_dummies(data_multicollinearity, drop_first=True)
```

```python
data_with_dummies.head()
```

```python
data_with_dummies.columns.values
```

```python
cols = ['log_price','Mileage', 'EngineV', 'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes']
```

```python
data_preprocessed = data_with_dummies[cols]
data_preprocessed.head()
```

```python
## Linear Regression Model
```

```python
from sklearn.linear_model import LinearRegression
```

```python
targets = data_preprocessed['log_price']
inputs = data_preprocessed.drop(['log_price'],axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(inputs)
```

```python
input_scaled = scaler.transform(inputs)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(input_scaled, targets, test_size=0.2, random_state=365)
```

```python
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
y_hat = reg.predict(x_train)
```

```python
plt.scatter(y_train, y_hat)
plt.xlabel("Targets(y_train)")
plt.ylabel("Predictions(y_hat)")
plt.show()
```

```python
sb.distplot(y_train-y_hat)
plt.title("Residuals PDF",size=12)
```

```python
reg.score(x_train,y_train)
```

```python
# This R squared value shows that our model explains 75% varaibility of the data
```

```python
reg.intercept_
```

```python
reg.coef_
```

```python
reg.summary = pd.DataFrame(inputs.columns.values, columns=["Features"])
reg.summary["weights"] = reg.coef_
reg.summary
```

```python
# Negative weights shows that price will decrease with that feature, 
# for positive weights price will increase with that feature
```

```python
# Positive weight for dummy variable shows the car is costly than Audi
# Negative weight shows car is cheaper than Audi
# As Audi was kept as base while creating dummies
```

```python
data_cleaned['Brand'].unique()
```

```python
y_hat_test = reg.predict(x_test)
```

```python
plt.scatter(y_test, y_hat_test,alpha=0.2)
plt.xlabel("Targets(y_test)")
plt.ylabel("Predictions(y_hat_test)")
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
df_pf = pd.DataFrame(np.exp(y_hat_test),columns=['Prediction'])
df_pf.head()
```

```python
df_pf['Target']= np.exp(y_test)
df_pf.head()
```

```python
y_test = y_test.reset_index(drop=True)
y_test.head()
```

```python
df_pf['Target']= np.exp(y_test)
df_pf.head()
```

```python
df_pf['Residual']= df_pf['Target']-df_pf['Prediction']
df_pf['Difference%']= np.absolute(df_pf['Residual']/df_pf['Target']*100)
```

```python
df_new=df_pf.sort_values(by=['Difference%'],ascending=False)
df_new
```
